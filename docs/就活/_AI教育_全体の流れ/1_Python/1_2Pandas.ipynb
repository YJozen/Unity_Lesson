# データの結合

```python
df_person1 = pd.read_csv("1_2/personNo_1.csv",index_col="社員番号")
df_person2 = pd.read_csv("1_2/personNo_2.csv",index_col="社員番号")
```

```python
df_person1
```

```python
df_person2 
```

## 1 joinによる結合

```python
df_person1.join(df_person2)# 社員番号をキーに結合 左外部結合　left joionがデフォルト
```

```python
df_person1.join(df_person2,how ="right")
```

```python
df_person1.join(df_person2,how ="inner") # あるやつだけ
```

```python
df_person = df_person1.join(df_person2,how ="outer")  #ないやつも
```

```python
df_person 
```

## 2 merge　途中のやつも見れる

```python
df_section = pd.read_csv("1_2/sectionNo.csv",index_col="組織番号")
```

```python
df_section
```

```python
pd.merge(df_person ,df_section,on="組織番号")
```

```python
pd.merge(df_person ,df_section,on=["組織番号","組織名"]) # 両方見て　あと　inner jooinがデフォルト
```

```python
pd.merge(df_person ,df_section,on=["組織番号","組織名"], how="outer")
```
