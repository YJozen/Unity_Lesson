# 5.　箱ひげ図

```python
df_score
```

## boxplot 箱ひげ図

```python
plt.figure(figsize=(10 ,8))
sns.boxplot(data=df_score,)# まる　外れ値
```

1組・2組・3組分ける

```python
df_score.melt()
```

```python
df_score.melt(["名前","クラス","性別"])#横にしたいものを指定　それ以外が縦に
```

```python
df_score_reshape = df_score.melt(["名前","クラス","性別"],var_name="科目",value_name="点数")
```

```python
df_score_reshape
```

```python
sns.boxplot(data=df_score_reshape,
            hue="クラス",
            x="科目",
            y="点数")
```

```python
plt.figure(figsize=(10 ,8))
sns.boxplot(data=df_score_reshape,
            hue="クラス",
            hue_order=["1組","2組","3組"],
            x="科目",
            y="点数")
```

violinplot バイオリンプロット

```python
sns.violinplot(data=df_score_reshape,
            hue="クラス",
            hue_order=["1組","2組","3組"],
            x="科目",
            y="点数")
```

どこに人が密集しているのかわかる

```python
plt.figure(figsize=(10 ,8))
sns.violinplot(data=df_score_reshape,
            hue="クラス",
            hue_order=["1組","2組","3組"],
            x="科目",
            y="点数")
```

## stripplot　散布図風ヒストグラム 点を重ねる

```python
plt.figure(figsize=(10 ,8))
sns.stripplot(data=df_score_reshape,
            hue="クラス",
            hue_order=["1組","2組","3組"],
            x="科目",
            y="点数")
```

## swarmplot　散布図風ヒストグラム2  点を綺麗に並べる

```python
plt.figure(figsize=(10 ,8))
sns.swarmplot(data=df_score_reshape,
            hue="クラス",
            hue_order=["1組","2組","3組"],
            x="科目",
            y="点数")
```

## heatmap ヒートマップ

相関係数

```python
df_score.corr(numeric_only=True)
```

```python
sns.heatmap(df_score.corr(numeric_only=True)) #何これ　見た目を整えよう
```

```python
sns.heatmap(df_score.corr(numeric_only=True),
            square=True,
            cmap="Blues",
            )
```

```python

sns.set(font='Hiragino Sans', context='talk',style='white')
plt.figure(figsize=(8 ,8))
sns.heatmap(df_score.corr(numeric_only=True),
            square=True,
            cmap="Blues",
            annot=True,
            )#数字を出す
```

```python
sns.heatmap(df_score.corr(numeric_only=True),
            square=True,
            cmap="Blues",
            annot=True,
            fmt=".2f",
            )
```

## clustermap 階層クラスタ付きヒートマップ

```python
sns.clustermap(df_score.corr(numeric_only=True),
            
            cmap   = "Blues",
            annot  = True,
            fmt    = ".2f",
            ) #  どれとどれが近いか　階層クラスタリング
```
